# Automatizacion de procesos en manejo de columnas

 ### Descripcion:
 
* Df = Es el DataFrame que se esta utilizando en este proceso.
* Var_name = el parametro al mismo tiempo el nombre de la columna que se desea cambiar.
* Dummy = variable reciclable donde se guardan los procesos del parametro var_name.

* La funcion drop() Elimina la columna.
  - En esta codificacion elimina y sobreescribe en el DataFrame.
* La funcion concat combina o mejor dicho concatena el DataFrame y la nueva columna.
  - En esta seccion concatena y sobreescribe en el DataFrame.
* El nombre de la funcion esta en mayuscula.

```python
def DummiesCreation(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy], axis = 1)
    return df
```

```python
# Demotracion!
DummiesCreation(data2, "sex")
```
